In this notebook we'll get familiar with our data.

```python
import os

data_path = os.path.join("/", "floyd", "input", "data")
print(data_path)

# Check the structure is the one we expect:
for directory in os.listdir(data_path):
    print(directory)
    data_subset_path = os.path.join(data_path, directory)
    for class_directory in os.listdir(data_subset_path):
        class_directory_path = os.path.join(data_subset_path, class_directory)
        number_of_images_in_class_directory = len(os.listdir(class_directory_path))
        print("\t - %s [%d images]" % (class_directory, number_of_images_in_class_directory))
```

## Exploring the photos

In the cell below we define a `show_random_photo` function that given a dataset and a label, will plot a random pictura in that subdomain. Feel free to use it to get acquainted with the data.

```python
import random

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline

def show_random_photo(dataset, label):
    assert dataset in {'test', 'valid', 'train'}, "dataset must be one of test, valid or train"
    assert label in {'seborrheic_keratosis', 'melanoma', 'nevus'}, "label must be one of seborrheic_keratosis, melanoma or nevus"
    
    path = os.path.join(data_path, dataset, label)
    
    random_photo = random.choice(os.listdir(path))
    photo_path = os.path.join(path, random_photo)

    img = mpimg.imread(photo_path)
    plt.imshow(img)
    plt.show()
    

DATASET = 'train'  # Chain accordingly
LABEL = 'seborrheic_keratosis'
show_random_photo(DATASET, LABEL)
```
